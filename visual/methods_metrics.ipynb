```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
path_acc_unet = 'model/trained/unet_acc.csv'
path_acc_deeplabv3plus = 'model/trained/deeplabv3plus_acc.csv'
path_acc_deeplabv3plus_mb2 = 'model/trained/deeplabv3plus_mobilev2_acc.csv'
acc_unet_df = pd.read_csv(path_acc_unet)
acc_deeplabv3plus_df = pd.read_csv(path_acc_deeplabv3plus)
acc_deeplabv3plus_mb2_df = pd.read_csv(path_acc_deeplabv3plus_mb2)
acc_unet_df.head()
```

```python
accs_df = [acc_unet_df, acc_deeplabv3plus_df, acc_deeplabv3plus_mb2_df]
colors = ['blue', 'orange', 'green']
labels = ['val_miou_unet', 
          'val_miou_deeplabv3plus', 
          'val_miou_deeplabv3plus_mb2']  
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, acc_df in enumerate(accs_df):
    sns.lineplot(data=acc_df, 
                x=acc_df.index, 
                y='val_miou', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
```

```python
fig, ax = plt.subplots(figsize=(6, 4))
for i, acc_df in enumerate(accs_df):
    sns.lineplot(data=acc_df, 
                x=acc_df.index, 
                y='tra_loss', 
                label=labels[i],
                color=colors[i],
                ax=ax)
ax.set_xlabel('Epochs')
# ax.set_ylim(0.6, 1)
```
